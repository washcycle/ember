```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("nyc-east-river-bicycle-crossings.zip")
df["Date"] = pd.to_datetime(df['Date'])
df["day_of_week"] = df["Date"].dt.day_name()
```

```python
df.head()
```

```python
day_of_week_2_int = {}
for dow_idx, dow in enumerate(pd.unique(df['day_of_week'])):
    day_of_week_2_int[dow] = dow_idx
```

```python
day_of_week_2_int
```

```python
df_encoded = df.copy()
df_encoded['day_of_week_encoded'] = df['day_of_week'].apply(lambda x: day_of_week_2_int[x])
```

```python
df_encoded.head()
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, Dropout, Input, Flatten
```

```python
embedding_size = 3
```

```python
def build_model():
    model = Sequential()
    model.add(Embedding(input_dim = 7+1, output_dim = embedding_size, input_length = 1, name="embedding"))
    model.add(Dense(units=40, activation = "relu"))
    model.add(Dropout(0.5))
    model.add(Dense(units=10, activation = "relu"))
    model.add(Dropout(0.5))
    model.add(Flatten())
    model.add(Dense(units=1, name="target"))
    
    return model
```

```python
model = build_model()
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer = "adam")
```

```python
X_train = np.reshape(df_encoded['day_of_week_encoded'].values, (-1, 1))        
```

```python
model.fit(x=X_train, y=np.reshape(df_encoded['Brooklyn Bridge'].values, (-1, 1)), 
          epochs = 200, 
          batch_size = 2)
```
